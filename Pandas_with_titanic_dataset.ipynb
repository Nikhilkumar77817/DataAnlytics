#Merge , Concate and Join

```python
# Merge two dataframes based on PassengerID
passenger_info = df[['Pclass','Gender','Age']].head(10).copy()
survival_info = df[['Survived']].head(10).copy()

# Add passenger ID for merge
passenger_info['PassengerID'] = range(1,11)
survival_info['PassengerID'] = range(1,11)

# Merge on PassengerID
merged_df = pd.merge(passenger_info, survival_info, on='PassengerID', how='inner')
merged_df
```

```python
# Create two DataFrames with index = PassengerID
df1 = df[['Age','Gender']].head(10).copy()
df1.index = range(1,11)

df2 = df[['Fare','Survived']].head(10).copy()
df2.index = range(1,11)

joined_df = df1.join(df2)
joined_df
```

```python
df['age_group'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
df[['Age','age_group']].head(10)
```

```python
df['fare_category'] = df['Fare'].apply(lambda x: 'Low' if x < 20 else ('Medium' if x < 50 else 'High'))
df[['Fare','fare_category']].head(10)
```

```python
df['family_size'] = df.apply(lambda row: row['SibSp'] + row['Parch'] + 1, axis=1)
df[['SibSp','Parch','family_size']].head(10)
```

```python
df['survival_label'] = df['Survived'].apply(lambda x: 'Survived' if x == 1 else 'Died')
df[['Survived','survival_label']].head(10)
```
