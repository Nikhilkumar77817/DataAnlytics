#**Seaborn**

Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

 Seaborn integrates tightly with Pandas DataFrames and works seamlessly with the rest of the PyData stack (NumPy, Pandas, Scipy). It is particularly well-suited for visualizing complex datasets with many variables.



# **Why Use Seaborn? (Advantages over Matplotlib)**

#While Matplotlib is highly customizable, it can be verbose for complex statistical plots. Seaborn addresses this with:

**Less Code, More Plot:** Creates sophisticated plots with just one or two lines of code.

**Attractive Defaults:** Comes with beautiful default styles and color palettes.

**Pandas Integration:** Works effortlessly with Pandas DataFrames. You can pass column names directly.

**Statistical Focus:** Many plots (like violin plots, box plots, and regression plots) are designed to reveal patterns in data.

**Easy Complex Plots:** Simplifies the creation of multi-plot grids for visualizing relationships across multiple variables.

**Built-in Color Palettes:** Provides a range of color palettes suitable for various data types (categorical, sequential, diverging).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#first method
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,2,8]
x1=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x1)
plt.show()
```

```python
#second method
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,2,8]

sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
#pair grid-  subplot grid  for plotting pairwise relationship dataset
graph= sns.PairGrid(data=penguins,hue="sex",  palette='Set2')
graph.map_upper(sns.scatterplot ) # upper triangle
graph.map_lower(sns.kdeplot ) # lower triangle
graph.map_diag(sns.histplot ) # diagonal
graph.add_legend()
```
