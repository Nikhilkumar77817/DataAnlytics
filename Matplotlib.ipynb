#**Matplotlib**

**Matplotlib** is a low level graph plotting library in python that serves as a visualization utility.

**Matplotlib** is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

**Basic Components**

**Figure:** The entire window or page that contains everything

**Axes:** The actual plot area where data is plotted

**Axis:** The x and y axes with ticks and labels

**Artist:** Everything visible on the figure (text, lines, etc.)

```python
# Generate sample data for the plot
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Calculate the sine of each x value

# Create a figure and an axes object for the plot
plt.figure(figsize=(6, 4))  # Create a figure with a specified size

# Plot the data
plt.plot(x, y, label='sin(x)', color='red', linewidth=2, linestyle='-')

# Add title and labels to the plot
plt.title('Sine Wave Function', fontsize=16, fontweight='bold')
plt.xlabel('X values', fontsize=12)
plt.ylabel('sin(X)', fontsize=12)

# Add a legend and a grid to the plot
plt.legend()
plt.grid(True, alpha=0.3) # Add a grid with some transparency

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

#**Sunplots** :

* By default, plt.plot() creates a single plot in one figure.

* But often we want to display multiple plots in the same figure (side by side or in a grid).

* For that, we use plt.subplot() or plt.subplots().

```python


x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [25,20,15,10,5]

plt.subplot(1,2,1)  # 1 row, 2 columns, first plot
plt.plot(x, y1, color="blue")
plt.title("Square Numbers")

plt.subplot(1,2,2)  # 1 row, 2 columns, second plot
plt.plot(x, y2, color="red")
plt.title("Descending Numbers")

plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.subplot(1,2,1) # 1 row,2 column,1 subplot
plt.plot(x,y)
plt.title('line chart')

plt.subplot(1,2,2)# 1 row,2 column,2 subplot
plt.bar(x,y)
plt.title('bar chart')
plt.tight_layout() # Subplots ke beech overlap avoid karta hai (very useful).
plt.show()
```

```python
# fig, ax = plt.subplots(nrows,ncols, figsize=(width,height))
fig, ax= plt.subplots(1,2,figsize=(10,5)) # Create a figure and a set of subplots
x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y) # Plot a line chart on the first subplot
ax[0].set_title('Line plot')

ax[1].bar(x,y) # Plot a bar chart on the second subplot
ax[1].set_title('bar plot')
plt.tight_layout()
plt.show()
```

```python
# plt.savefig("filename.format",dpi =value,bbox_inches='tighti)  dpi=dot par inch used for image resolution
x=[1,2,3,4]
y=[10,20,15,25]
plt.plot(x,y,color='orange',marker='o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('simple line plot')
plt.grid(True)

plt.savefig("lineplot.png",dpi =300,bbox_inches='tight') # save the figure
plt.show()
```

#Advance concept of matplotlib

```python
# Create 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

# Top-left: Line plot
x = np.linspace(0, 10, 100) # Generate 100 points from 0 to 10
axes[0, 0].plot(x, np.sin(x), 'b-') # Plot sine wave
axes[0, 0].set_title('Sine Wave') # Set title for subplot

# Top-right: Scatter plot
x_scatter = np.random.rand(50) # Generate 50 random x values
y_scatter = np.random.rand(50) # Generate 50 random y values
axes[0, 1].scatter(x_scatter, y_scatter, c=x_scatter, cmap='viridis') # Create scatter plot
axes[0, 1].set_title('Scatter Plot') # Set title for subplot

# Bottom-left: Bar plot
categories = ['A', 'B', 'C'] # Define categories
values = [10, 20, 15] # Define values
axes[1, 0].bar(categories, values, color=['red', 'green', 'blue']) # Create bar plot
axes[1, 0].set_title('Bar Plot') # Set title for subplot

# Bottom-right: Histogram
data = np.random.normal(0, 1, 1000) # Generate 1000 random data points from a normal distribution
axes[1, 1].hist(data, bins=30, alpha=0.7, color='orange') # Create histogram with 30 bins
axes[1, 1].set_title('Histogram') # Set title for subplot

plt.tight_layout() # Adjust layout to prevent titles/labels from overlapping
plt.show() # Display the plots
```

```python
# Advanced customization
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(8, 4))

# Plot with custom styling
ax.plot(x, y1, label='sin(x)', color='navy', linewidth=3, linestyle='-',
        marker='o', markersize=4, markevery=10) # Plot the first line with custom styling
ax.plot(x, y2, label='cos(x)', color='crimson', linewidth=2, linestyle='--') # Plot the second line with custom styling

# Customize axes
ax.set_xlim(0, 10) # Set the limits of the x-axis
ax.set_ylim(-1.5, 1.5) # Set the limits of the y-axis
ax.set_xlabel('X Axis', fontsize=14, fontweight='bold') # Set the label for the x-axis
ax.set_ylabel('Y Axis', fontsize=14, fontweight='bold') # Set the label for the y-axis
ax.set_title('Customized Plot', fontsize=16, fontweight='bold') # Set the title of the plot

# Customize grid and ticks
ax.grid(True, linestyle='--', alpha=0.7) # Add a grid with custom styling
ax.set_xticks(np.arange(0, 11, 2)) # Set the tick marks on the x-axis
ax.set_yticks(np.arange(-1.5, 2.0, 0.5)) # Set the tick marks on the y-axis

# Add legend
ax.legend(loc='upper right', fontsize=12, frameon=True, fancybox=True, shadow=True) # Add a legend with custom styling

# Add text annotation
ax.text(2, 0.5, 'Maximum Area', fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5)) # Add text annotation to the plot

# Add horizontal and vertical lines
ax.axhline(y=0, color='black', linestyle='-', alpha=0.3) # Add a horizontal line
ax.axvline(x=5, color='gray', linestyle=':', alpha=0.5) # Add a vertical line

plt.tight_layout() # Adjust layout to prevent titles/labels from overlapping
plt.show() # Display the plot
```

```python
# Error bar example
x = np.arange(1, 6) # Create an array of numbers from 1 to 5
y = np.array([10, 15, 13, 18, 16]) # Create an array of y-values
y_err = np.array([1, 2, 1.5, 3, 2]) # Create an array of error values for each y-value

fig, ax = plt.subplots(figsize=(10, 6)) # Create a figure and an axes object with a specified size
ax.errorbar(x, y, yerr=y_err, fmt='o', color='blue', # Plot the error bars with x, y, and error values
            ecolor='red', elinewidth=2, capsize=5, capthick=2) # Customize the appearance of the error bars

ax.set_title('Error Bar Plot') # Set the title of the plot
ax.set_xlabel('Measurement') # Set the label for the x-axis
ax.set_ylabel('Value  Error') # Set the label for the y-axis
ax.set_xticks(x) # Set the tick marks on the x-axis to match the x-values
plt.show() # Display the plot
```

```python
# Customize global settings
plt.rcParams.update({
    'font.size': 12,
    'font.family': 'serif',
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'figure.figsize': (10, 6),
    'grid.alpha': 0.3,
    'lines.linewidth': 2,
})

# Create plot with custom style
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')
ax.legend()
ax.grid(True)
ax.set_title('Custom Styled Plot')
plt.show()
```

**What is a Boxplot?**

A boxplot (also called whisker plot) is a way to show how data is spread out.
It displays:

Minimum

First Quartile (Q1 = 25th percentile)

Median (Q2 = 50th percentile)

Third Quartile (Q3 = 75th percentile)

Maximum

Outliers (values that are far away from the rest)



```python

# data
math_scores = [55, 60, 62, 65, 70, 75, 80, 85, 90, 95]
science_scores = [50, 52, 55, 58, 60, 62, 64, 66, 68, 95]

# boxplot of two subjects
plt.boxplot([math_scores, science_scores], labels=["Math", "Science"])
plt.title("Students' Marks in Math vs Science")
plt.ylabel("Scores")
plt.show()
```

```python
dept_a = [25, 28, 30, 35, 40, 42, 45, 47, 50, 55]
dept_b = [30, 32, 34, 36, 38, 40, 42, 44, 70, 80]
dept_c = [20, 22, 24, 24, 25, 26, 26, 27, 28, 30]

box = plt.boxplot([dept_a, dept_b, dept_c],
                  labels=["Dept A", "Dept B", "Dept C"],
                  patch_artist=True)

colors = ['pink', 'lightyellow', 'lightcyan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.title("Employee Salaries by Department")
plt.ylabel("Salary (in thousands)")
plt.show()
```
